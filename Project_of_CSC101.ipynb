```python
import math

accounts = []  # two dimensional array to store account information: Account name, number and balance
accountNo = 1000  # account number for the first ever account created


def addAccount():
    '''
    This function creates and adds a new account to the accounts list
    '''
    global accountNo  # access the global variable to make changes to it
    name = input("Enter Account Name:")
    amount = int(input("Enter Amount:"))
    account = [name, accountNo, amount]
    accounts.append(account)  # add new account information to the accounts list
    accountNo += 1  # incerement account number for the next new account


def accountBalance():
    '''
    Shows the account balance according to account number
    '''
    acNo = int(input("Account No:"))
    acFound = False
    for i in accounts:  # loop through each account to match account number
        if(i[1] == acNo):  # if matching account number has been found
            print("Balance :", i[2])
            acFound = True
            break
    if(acFound == False):
        print("Account Not Found !")


def depositeMoney():
    '''
    Adds money to existing account balance
    '''
    acNo = int(input("Account No:"))
    acFound = False
    for i in accounts:
        if(i[1] == acNo):
            acFound = True
            amount = float(input("Enter Amount :"))
            i[2] += amount
            break
    if(acFound == False):
        print("Account Not Found !")


def withdrawMoney():
    '''
    Subtracts amount entered by user from account balance if amount does not exceed account balance
    '''
    acNo = int(input("Account No:"))
    acFound = False
    for i in accounts:
        if(i[1] == acNo):
            acFound = True
            while(True):
                amount = float(input("Enter Amount :"))
                if(amount >= i[2] or amount < 0):
                    print("Not Sufficient Amount")
                    print("Enter 0 To exit")
                    continue
                else:
                    break
            i[2] = i[2]-amount
            break
    if(acFound == False):  # if none of the accounts match the entered account number
        print("Account Not Found !")


def showAccounts():
    '''
    Displays information of all accounts in accounts list
    '''
    for i in accounts:
        print()
        print('Account Name:', i[0])
        print('Account Number:', i[1])
        print('Account Balance:', i[2])


def internRate(n: int):
    '''
    Calculates and returns interest rate based on account balance
    '''
    if(1000 <= n < 5000):
        return 3
    elif(5000 <= n <= 10000):
        return 3.25
    elif(10000 <= n < 50000):
        return 3.55
    elif(50000 <= n < 100000):
        return 3.8
    elif(10000 <= n):
        return 4
    else:
        return 0


def checkInterest():
    acNo = int(input("Account No:"))
    acFound = False
    for i in accounts:
        # if account has been found according to account number
        if(i[1] == acNo):
            rate = internRate(i[2])
            print('Interest Rate:', rate)
            rate /= 100
            rate += 1
            # math.ceil is used to round up the interest rate value
            print('Balance After 1 Month:', math.ceil(rate*i[2]))
            print('Balance After 6 Months:', math.ceil((rate**6)*i[2]))
            print('Balance After 1 Year:', math.ceil((rate**12)*i[2]))
            print('Balance After 5 Years:', math.ceil((rate**60)*i[2]))
            print('Balance After 10 Years:', math.ceil((rate**120)*i[2]))
            print()
            break
    if(acFound == False):
        print("Account Not Found !")


def accountBalance():
    acNo = int(input("Account No:"))
    acFound = False
    for i in accounts:
        if(i[1] == acNo):
            print("Balance :", i[2])
            break
    if(acFound == False):
        print("Account Not Found !")


while(True):
    print("0.Exit")
    print("1.Show All Accounts")
    print("2.Create New Account")
    print("3.Deposite Money")
    print("4.Withdraw Money")
    print("5.Check Balance")
    print("6.Check Interest")
    choice = int(input("Enter Your choice : "))
    if(choice == 1):
        showAccounts()
    elif(choice == 2):
        addAccount()
    elif(choice == 3):
        depositeMoney()
    elif(choice == 4):
        withdrawMoney()
    elif(choice == 5):
        accountBalance()
    elif (choice == 6):
        checkInterest()
    elif (choice == 0):
        break
```
